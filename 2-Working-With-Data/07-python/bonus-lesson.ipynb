## Series

### Series is like a list or 1D-array, but with index.It is generally used to represent data associated with time.

```r
stock_prices = c(799,1174.8,865.1,1334.6,635.4,918.5,685.5,998.6,784.2,985,882.8,1071)
stock_prices_glennmark = ts(stock_prices,start = c(2020,1,1),frequency = 12)
stock_prices_glennmark
plot(stock_prices_glennmark)
```

```r
# CREATING A TIME SERIES
randomData<- rnorm(100)
month <- ts(randomData,start=c(2020,1),frequency=12)
plot(month)
## it bit much longer a it is 100/4 takes much longer year
```

**DataFrame**
#### It is a two dimensional data structure in which data is inserted in tabular form.

```r
STOCK_PRICE <- data.frame(
USD=c(1.02,1.085,2.05,3.45,4.05,5.32,3.45,4.65,6.65,9.23,10.25,12.69),
EURO = c(0.85,1.02,1.85,3.02,3.35,4.35,2.65,3.95,5.55,8.90,9.80,11.50))
STOCK_PRICES  <- ts(STOCK_PRICE,start = c(2020,1,1),frequency = 12)
```

## Grouping 

### group_by function  group rows of data and frame with respect to one or more variabes

```r
library(dplyr)
```

```r
df1 <- group_by(STOCK_PRICE,EURO)
```

```r
summarise(df1,mymean =mean(EURO))
```

```r
df2<-group_by(df1,USD>3.2)
```

```r
summarise(df2,mymean = mean(USD))
```
